```python
#To change columns from string to numbers, remove the string characters.
amazon['Close/Last in $'] = amazon['Close/Last'].str.replace('$', '')
apple['Close/Last in $'] = apple['Close/Last'].str.replace('$', '')
meta['Close/Last in $'] = meta['Close/Last'].str.replace('$', '')
microsoft['Close/Last in $'] = microsoft['Close/Last'].str.replace('$', '')
netflix['Close/Last in $'] = netflix['Close/Last'].str.replace('$', '')
tesla['Close/Last in $'] = tesla['Close/Last'].str.replace('$', '')
```

```python
amazon['Open in $'] = amazon['Open'].str.replace('$', '')
apple['Open in $'] = apple['Open'].str.replace('$', '')
meta['Open in $'] = meta['Open'].str.replace('$', '')
microsoft['Open in $'] = microsoft['Open'].str.replace('$', '')
netflix['Open in $'] = netflix['Open'].str.replace('$', '')
tesla['Open in $'] = tesla['Open'].str.replace('$', '')
```

```python
amazon['High in $'] = amazon['High'].str.replace('$', '')
apple['High in $'] = apple['High'].str.replace('$', '')
meta['High in $'] = meta['High'].str.replace('$', '')
microsoft['High in $'] = microsoft['High'].str.replace('$', '')
netflix['High in $'] = netflix['High'].str.replace('$', '')
tesla['High in $'] = tesla['High'].str.replace('$', '')
```

```python
amazon['Low in $'] = amazon['Low'].str.replace('$', '')
apple['Low in $'] = apple['Low'].str.replace('$', '')
meta['Low in $'] = meta['Low'].str.replace('$', '')
microsoft['Low in $'] = microsoft['Low'].str.replace('$', '')
netflix['Low in $'] = netflix['Low'].str.replace('$', '')
tesla['Low in $'] = tesla['Low'].str.replace('$', '')
```

```python
# Dropping the Old Columns: Close/Last, Open, High','Low'
amazon.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
apple.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
meta.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
microsoft.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
netflix.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
tesla.drop(columns = ['Close/Last','Open','High','Low'], inplace = True)
```

Step 5: Converting values of columns to appropriate datatype

```python
# Convert the datatype of Date column from object to Datetime.
amazon['Date'] = pd.to_datetime(amazon['Date'])
apple['Date'] = pd.to_datetime(apple['Date'])
meta['Date'] = pd.to_datetime(meta['Date'])
microsoft['Date'] = pd.to_datetime(microsoft['Date'])
netflix['Date'] = pd.to_datetime(netflix['Date'])
tesla['Date'] = pd.to_datetime(tesla['Date'])
```

```python
# Convert the datatype of columns from object to float type.
amazon['Close/Last in $'] = amazon['Close/Last in $'].astype(float).round(4)
apple['Close/Last in $'] = apple['Close/Last in $'].astype(float).round(4)
meta['Close/Last in $'] = meta['Close/Last in $'].astype(float).round(4)
microsoft['Close/Last in $'] = microsoft['Close/Last in $'].astype(float).round(4)
netflix['Close/Last in $'] = netflix['Close/Last in $'].astype(float).round(4)
tesla['Close/Last in $'] = tesla['Close/Last in $'].astype(float).round(4)
```

```python
amazon['Open in $'] = amazon['Open in $'].astype(float).round(4)
apple['Open in $'] = apple['Open in $'].astype(float).round(4)
meta['Open in $'] = meta['Open in $'].astype(float).round(4)
microsoft['Open in $'] = microsoft['Open in $'].astype(float).round(4)
netflix['Open in $'] = netflix['Open in $'].astype(float).round(4)
tesla['Open in $'] = tesla['Open in $'].astype(float).round(4)
```

```python
amazon['High in $'] = amazon['High in $'].astype(float).round(4)
apple['High in $'] = apple['High in $'].astype(float).round(4)
meta['High in $'] = meta['High in $'].astype(float).round(4)
microsoft['High in $'] = microsoft['High in $'].astype(float).round(4)
netflix['High in $'] = netflix['High in $'].astype(float).round(4)
tesla['High in $'] = tesla['High in $'].astype(float).round(4)
```

```python
amazon['Low in $'] = amazon['High in $'].astype(float).round(4)
apple['Low in $'] = apple['High in $'].astype(float).round(4)
meta['High in $'] = meta['High in $'].astype(float).round(4)
microsoft['High in $'] = microsoft['High in $'].astype(float).round(4)
netflix['High in $'] = netflix['High in $'].astype(float).round(4)
tesla['High in $'] = tesla['High in $'].astype(float).round(4)
```

```python
#To prep for moving averages and rolling over function, sort date column chronically per dataset
amazon.sort_values(by=['Date'], inplace=True) 
apple.sort_values(by=['Date'], inplace=True) 
meta.sort_values(by=['Date'], inplace=True) 
microsoft.sort_values(by=['Date'], inplace=True) 
netflix.sort_values(by=['Date'], inplace=True) 
tesla.sort_values(by=['Date'], inplace=True) 
```

```python
# Inserting columns to each dataset where it points out the 6 month and one year moving average (in days)
```

```python
amazon['MA_Six_Months'] = amazon['Close/Last in $'].rolling(window=183).mean()
amazon['MA_One_Year'] = amazon['Close/Last in $'].rolling(window=365).mean()
amazon
```

```python
microsoft['MA_Six_Months'] = microsoft['Close/Last in $'].rolling(window=183).mean()
microsoft['MA_One_Year'] = microsoft['Close/Last in $'].rolling(window=365).mean()
microsoft
```

```python
netflix['MA_Six_Months'] = netflix['Close/Last in $'].rolling(window=183).mean()
netflix['MA_One_Year'] = netflix['Close/Last in $'].rolling(window=365).mean()
netflix
```

```python
tesla['MA_Six_Months'] = tesla['Close/Last in $'].rolling(window=183).mean()
tesla['MA_One_Year'] = tesla['Close/Last in $'].rolling(window=365).mean()
tesla
```
